# Cleaning the data

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("C:/Users/sriji/OneDrive/Desktop/CO2 emission.csv")
```

```python
data.drop(['MODELYEAR','MODEL','VEHICLECLASS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY'],axis=1, inplace=True)
```

```python
print(data.head())
```

```python
data.isnull()
```

```python
data.FUELTYPE.unique()
```

# Finding central tendencies

```python
#FUELCONSUMPTION_COMB_MPG

print('Max: ',data.max()['FUELCONSUMPTION_COMB_MPG'])
print('Mode: ',data.mode()['FUELCONSUMPTION_COMB_MPG'])
print('Mean: ',data['FUELCONSUMPTION_COMB_MPG'].mean())
print('Median: ',data['FUELCONSUMPTION_COMB_MPG'].median())
```

```python
#FUELCONSUMPTION_COMB

print('Max: ',data.max()['FUELCONSUMPTION_COMB'])
print('Mode: ',data.mode()['FUELCONSUMPTION_COMB'])
print('Mean: ',data['FUELCONSUMPTION_COMB'].mean())
print('Median: ',data['FUELCONSUMPTION_COMB'].median())
```

```python
#CO2EMISSIONS

print('Max: ',data.max()['CO2EMISSIONS'])
print('Mode: ',data.mode()['CO2EMISSIONS'])
print('Mean: ',data['CO2EMISSIONS'].mean())
print('Median: ',data['CO2EMISSIONS'].median())
```

# Visualising the data

```python
plt.style.use('bmh')
```

```python
#BAR GRAPH

x=data['MAKE']
y=data['FUELCONSUMPTION_COMB']

plt.xlabel('MAKE', fontsize=10)
plt.ylabel('FUELCONSUMPTION_COMB', fontsize=10)

plt.bar(x,y)

plt.show()
```
